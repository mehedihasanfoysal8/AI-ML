*Numpy*

```python
import numpy as np

# 1d Array --
arr1 = np.array([1, 2, 3, 4, 5])

# print(arr1.ndim, end=" ")

# 2d Array --
arr2 = np.array([[1, 3, 6, 9], [2, 4, 6, 8]])

print(arr2)
```

*Array Attibutes*

```python
# Dimenstion
print(arr1.ndim)

# shape
print(arr1.shape)
print(arr2.shape)

# D-Type
print(arr1.dtype)
```

*NumPy Data Type*

```python
arr = np.array([1, 2, 3])

# upcusting int to float
print(arr.dtype)

# upcasting int to float

arr1 = np.array([1, 2 , 2.5])

print(arr1.dtype)
```

*Select a data type for an array*

```python
arr = np.array([1, 2, 4, 6], dtype=np.int8)
print(arr)
print(arr.dtype)
arr = arr.astype(np.int16)
```

```python
matrix = np.array([[4, 5, 8], [6, 7, 2]])

print(matrix.dtype)
print(matrix.shape)
```

```python
sets = {4, 7, 8, 9, 8}

arr = np.array(list(sets)%%!)
print(arr)
```

*Random Values*

```python
arr = np.random.rand(10, 5)
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python
# np.random.randint(start, end, shape)

arr = np.random.randint(10, 50, (5, 5))

print(arr)
```

```python
# np.random.uniform(start, end, shape)

arr = np.random.uniform(10, 50, (5, 5))

print(arr)
```

*Arange Function*

```python
# arange accept(where start, where end, and step) --- simpiler to loop range function
arr = np.arange(50, 101, 5)

print(arr)
```

```python
arr = np.array([1, 2, 3, 4, 5])
arr[1:4] = 0
print(arr)
```
