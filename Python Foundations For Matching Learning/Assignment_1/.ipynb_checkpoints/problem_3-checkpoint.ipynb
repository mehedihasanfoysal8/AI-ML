```python
import string

with open("data.txt", 'r') as file:
    text = file.read()

text = text.lower()

for punctuation in string.punctuation:
    text = text.replace(punctuation, "")

words = text.split()

mx_cnt = 0
max_word = ""

for word in words:
    cnt = words.count(word)
    
    if cnt > mx_cnt:
        mx_cnt = cnt
        max_word = word

print(max_word)
```
