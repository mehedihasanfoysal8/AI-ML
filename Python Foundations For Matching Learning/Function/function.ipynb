```python
def return_doc(a, b):
    """
    Adds two integers and returns the result.

    Parameters:
    a (int): First number
    b (int): Second number

    Returns:
    int: Sum of a and b
    """
    result = a + b
    return result
```

```python
def adder(a, b, c, d):
    print(f" {a} {b} {c} {d}")

adder(10, 20, 30, 40)
```

```python
def adder(*args):
    sum = 0
    for i in args:
        sum = sum + i
    
    print(f'Final resutl is {sum}')

adder(10, 20, 30, 40, 50)
```
