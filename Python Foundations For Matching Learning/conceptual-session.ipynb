```python
arr = [1, 2, 3, 4, 5, 6]

def fun1(x):
    return x * 10
    
# here use user define function
arr2 = list(map(fun1, arr))
print(arr2)


# here use lambda function
arr3 = list(map(lambda x : x * 10, arr))

print(arr3)


# This is manula way to muliplay 
#for i in arr:
    #print(fun1(i))
```

```python
arr = [80, 70, 40, 50, 60]

def result(x):
    if x >= 80:
        return f"Result is {x} A+"
    elif x >= 70:
        return f"Result is {x} A-"
    elif x >= 60:
        return f"Result is {x} B"
    else:
        return f"Result is {x} F"

res = list(map(result, arr))
print(res)
```

```python
arr = [23, 56, 77, 443, 80, 90]

result = list(filter(lambda x : x >= 60, arr))

print(result)
```

```python
from functools import reduce

arr = [1, 2, 3, 4, 5, 6]

result = reduce( lambda x, y : x + y, arr)

print(result)
```
