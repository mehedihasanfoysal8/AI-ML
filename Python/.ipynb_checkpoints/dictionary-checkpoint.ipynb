Dictionary হলো key-value pair data structure

```python
student = {
    "name": "Foysal",
    "age": 20,
    "city": "Dhaka"
}
```

```python
student = {"name": "Foysal", "age": 20}

print(student["name"])  # Foysal
```

get() method (safe way)

```python
print(student["age"] = 25
print(student).get("age"))        # 20
print(student.get("email"))      # None (error দেয় না)
```

```python
student["age"] = 25
print(student)
```

New key add করা

```python
student["email"] = "f@gmail.com"
```

```python
student = {
    "name": "Foysal",
    "age": 20
}

# add
student["city"] = "Dhaka"

# update
student.update({"age": 22, "email": "f@gmail.com"})

# access
print(student["name"])

# keys
print(student.keys())

# values
print(student.values())

# items
print(student.items())

# delete
student.pop("age")

print(student)
```
