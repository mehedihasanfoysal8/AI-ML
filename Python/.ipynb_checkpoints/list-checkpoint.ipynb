```python
nums = [10, 20, 30]

# append
nums.append(40)
print(nums)   # [10, 20, 30, 40]

# insert
nums.insert(1, 15)
print(nums)   # [10, 15, 20, 30, 40]

# extend
nums.extend([50, 60])
print(nums)   # [10, 15, 20, 30, 40, 50, 60]

# remove
nums.remove(20)
print(nums)   # [10, 15, 30, 40, 50, 60]

# pop
nums.pop(2)
print(nums)   # [10, 15, 40, 50, 60]

# index
print(nums.index(40))  # 2

# count
print(nums.count(10))  # 1

# sort
nums.sort()
print(nums)   # sorted list

# reverse
nums.reverse()
print(nums)

# copy
new_list = nums.copy()
print(new_list)

# clear
nums.clear()
print(nums)   # []
```

```python
random_list = [ x for x in range(2, 20) if x % 2 != 0]
random_list
```
