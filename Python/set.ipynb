```python
a = {1, 2, 3}
b = {3, 4, 5}

# add
a.add(10)
print(a)

# update
a.update([20, 30])
print(a)

# remove
a.remove(2)
print(a)

# discard
a.discard(100)  # error না

# union
print(a.union(b))

# intersection
print(a.intersection(b))

# difference
print(a.difference(b))

# subset
print({1, 3}.issubset(a))

# clear
a.clear()
print(a)
```
